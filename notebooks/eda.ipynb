```python
import pandas as pd

df = pd.read_csv('../data/raw.csv')
df.head(1000)
```

```python
# Obtener fecha de inicio y fin
df['date'] = pd.to_datetime(df['date'])
df['date'].min(), df['date'].max()
```

```python
# Obtener a partir de que "date" hay "time" != NaN
df_notna = df[df['time'].notna()]
df_notna['date'].min()
```

```python
# Dataframe con "tweet"'s repetidos
df_repeated = df[df.duplicated(subset=['tweet'], keep=False)]

# Mostrar valores repetidos
df_repeated.sort_values(by='tweet')
```
